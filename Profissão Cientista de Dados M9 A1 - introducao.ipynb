# 6.0 Group by DTNASC e contagem da IDADEMAE <a name="dtnasc"></a>

##  6.1 Em SQL <a name="sql"></a>

``` sql
SELECT DTNASC, count(IDADEMAE) 
FROM sinasc
GROUP BY DTNASC

```

## 6.3 Python com ``groupby()`` <a name="py_group"></a>

![Esquema de Group By usando Excel](Profissão_Cientista_de_Dados_M9_A1_groupby_gif_2.gif)

````python
sinasc_1.groupby('DTNASC')['IDADEMAE'].count().to_frame().reset_index()
````

## 6.4 Python com ``pivot_table()`` <a name="py_pivot"></a>

````python
sinasc_1.pivot_table(values = 'IDADEMAE',index = 'DTNASC', aggfunc='count').reset_index()
````

````python
sinasc_1.pivot_table(values = 'IDADEMAE',index = 'DTNASC', aggfunc='mean').plot(figsize=[18,5])
plt.title('Média das idades das mães por data de nascimento')
plt.xlabel('Data de nascimento')
plt.ylabel('Média das idades das mães')
plt.grid(True)
plt.show()
````
